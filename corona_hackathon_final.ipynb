# CLEANING

```python
def remove_hashtag_link(text):
    text  = " ".join([char for char in text.split(" ") if not char.startswith("#") and not char.startswith("http") and not char.lower() in stop_words])
    return text
```

```python
tweet_df['Tweet'] = tweet_df['text'].apply(lambda x: remove_hashtag_link(x))
```

```python
tweet_df[["text","Tweet"]].head(5)
```

explaination of re sub:(starting with @)or(anything that does not start with alpha numeric

i.e If the first character of the set is '^', all the characters that are not in the set will be matched-so special chars will be removed)or (words before : and spaces) and punctuation removed

```python
def remove_punct(text):
    text = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())
    return text
```

```python
tweet_df['Tweet'] = tweet_df['Tweet'].apply(lambda x: remove_punct(x)).apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
```

# Topic Extraction from Tweets using NMF

Now that we have clean text we can apply some processing to turn the clean tweets into vectors and then build a model.

```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(min_df=5, analyzer='word', ngram_range=(1, 2), stop_words=stop_words.append(["get"]))
vz = vectorizer.fit_transform(tweet_df['Tweet'])

nmf = NMF(n_components=5, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(vz)

feature_names = vectorizer.get_feature_names()
no_top_words = 10
word_dict = {}

for topic_idx, topic in enumerate(nmf.components_[:10]):
    word_dict["Topic %d:"% (topic_idx)] = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]

pd.DataFrame.from_dict(word_dict, orient='index')
```
